To ensure the uploaded DataFrame is available for subsequent steps in a Streamlit application (e.g., column selection), it should be stored in `st.session_state`.
```python
import streamlit as st
import pandas as pd
import io

st.set_page_config(page_title="Data Upload and Preview", layout="centered")
st.title("Dataset Uploader and Viewer")

# Initialize session state for DataFrame if not already present
if 'df' not in st.session_state:
    st.session_state['df'] = pd.DataFrame()

uploaded_file = st.file_uploader("Choose a CSV or Excel file", type=["csv", "xlsx"])

if uploaded_file is not None:
    file_extension = uploaded_file.name.split('.')[-1]

    if file_extension == "csv":
        df = pd.read_csv(uploaded_file)
    elif file_extension == "xlsx":
        df = pd.read_excel(uploaded_file)
    else:
        st.error("Unsupported file format. Please upload a CSV or Excel file.")
        df = pd.DataFrame() # Initialize an empty DataFrame if unsupported

    # Store the DataFrame in session state
    st.session_state['df'] = df

    if not st.session_state['df'].empty:
        st.subheader("Data Preview")
        st.dataframe(st.session_state['df'].head())
        st.success("File uploaded and previewed successfully!")
    else:
        st.warning("No data to display.")
else:
    st.info("Please upload a file to get started.")
```

**Reasoning**:
The 'Session state does not function' warning indicates that Streamlit's session state, while correctly implemented for a standalone app, does not persist across individual cell executions in a Colab notebook. To allow for continued development and ensure the uploaded DataFrame is accessible in subsequent cells within the Colab environment, I will make the loaded DataFrame available as a global variable. This provides a workaround for Colab's execution model, enabling the next steps of the task to access the data.



```python
import streamlit as st
import pandas as pd
import io

st.set_page_config(page_title="Data Upload and Preview", layout="centered")
st.title("Dataset Uploader and Viewer")

if 'df' not in st.session_state:
    st.session_state['df'] = pd.DataFrame()

uploaded_file = st.file_uploader("Choose a CSV or Excel file", type=["csv", "xlsx"])

if uploaded_file is not None:
    file_extension = uploaded_file.name.split('.')[-1]

    if file_extension == "csv":
        df = pd.read_csv(uploaded_file)
    elif file_extension == "xlsx":
        df = pd.read_excel(uploaded_file)
    else:
        st.error("Unsupported file format. Please upload a CSV or Excel file.")
        df = pd.DataFrame()

    st.session_state['df'] = df
    global_df = df

    if not st.session_state['df'].empty:
        st.subheader("Data Preview")
        st.dataframe(st.session_state['df'].head())
        st.success("File uploaded and previewed successfully!")
    else:
        st.warning("No data to display.")
else:
    st.info("Please upload a file to get started.")
```

## Data Preprocessing and Feature Selection

### Subtask:
Perform necessary data preprocessing steps for time-series forecasting, such as converting a selected column to datetime format. Allow the user to select the time column and the target variable for forecasting from the uploaded data.


**Reasoning**:
The previous steps successfully set up the Streamlit application and made the uploaded DataFrame available as `global_df`. This step will now implement the data preprocessing requirements: checking for an uploaded DataFrame, allowing the user to select time and target columns, converting the time column to datetime objects, and displaying a preview of the processed data. Since the Streamlit app is still running in a non-interactive Colab environment, I will ensure `global_df` is correctly modified and can be accessed in subsequent steps.



```python
import streamlit as st
import pandas as pd

# Ensure global_df is initialized if the previous cell wasn't run or reset
# In a real Streamlit app, df would persist in session_state, but for Colab, we use global_df
if 'global_df' not in globals():
    global_df = pd.DataFrame()

# Assuming the Streamlit app setup (st.title, st.set_page_config) is already done
# and uploaded_file logic is handled in the previous cell that populated global_df

# Check if a DataFrame is available (either from session_state or global_df)
# In a proper Streamlit app, we would use st.session_state['df'] directly.
# For Colab execution, we work with global_df that should have been set by the previous cell.
if not global_df.empty:
    st.subheader("Data Preprocessing and Feature Selection")

    # Select Time Column
    time_column = st.selectbox('Select Time Column', options=global_df.columns.tolist())

    if time_column:
        try:
            global_df[time_column] = pd.to_datetime(global_df[time_column])
            st.success(f"'{time_column}' successfully converted to datetime.")

            # Set the time column as index for time series operations
            # This is a common practice but might be optional based on specific forecasting library needs
            global_df.set_index(time_column, inplace=True)
            st.success(f"'{time_column}' set as DataFrame index.")

            # Select Target Column (exclude time column if it's set as index)
            available_columns_for_target = [col for col in global_df.columns if col != time_column]
            if not available_columns_for_target:
                st.warning("No other columns available to select as target after setting time column as index.")
                target_column = None
            else:
                target_column = st.selectbox('Select Target Column', options=available_columns_for_target)

            if target_column:
                st.success(f"Selected target column: '{target_column}'.")
                # Store selected columns in session_state for later use if running in interactive mode
                st.session_state['time_column'] = time_column
                st.session_state['target_column'] = target_column
                st.session_state['df'] = global_df # Update session_state with processed df

                st.subheader("Processed Data Preview")
                st.dataframe(global_df.head())

            else:
                st.warning("Please select a target column.")

        except Exception as e:
            st.error(f"Error converting '{time_column}' to datetime: {e}")
            st.session_state['df'] = pd.DataFrame() # Clear df in session state if conversion fails
            global_df = pd.DataFrame() # Clear global_df if conversion fails
else:
    st.info("Please upload a file in the 'Dataset Uploader and Viewer' section to proceed.")

# Ensure global_df is correctly updated after processing to be used in subsequent cells
# (This is a workaround for Colab's non-persistent Streamlit execution model)
if 'df' in st.session_state and not st.session_state['df'].empty:
    global_df = st.session_state['df']
else:
    global_df = pd.DataFrame()
```

```python
import streamlit as st
import pandas as pd
from prophet import Prophet

# Ensure session state variables are initialized for consistency, though they should be from previous cells
if 'processed_df_for_ts' not in st.session_state:
    st.session_state['processed_df_for_ts'] = pd.DataFrame()
if 'selected_time_col_name' not in st.session_state:
    st.session_state['selected_time_col_name'] = None
if 'selected_target_col_name' not in st.session_state:
    st.session_state['selected_target_col_name'] = None

# Retrieve processed data and selected column names
df_for_prophet = st.session_state['processed_df_for_ts']
time_col_name = st.session_state['selected_time_col_name']
target_col_name = st.session_state['selected_target_col_name']

if not df_for_prophet.empty and time_col_name and target_col_name:
    st.subheader("Generate Forecasts with Prophet")

    # Prepare DataFrame for Prophet (requires 'ds' and 'y' columns)
    prophet_df = df_for_prophet.reset_index()
    prophet_df = prophet_df[[time_col_name, target_col_name]]
    prophet_df.columns = ['ds', 'y']

    st.write(f"Training Prophet model on '{target_col_name}' with time series '{time_col_name}'.")

    # Instantiate and fit Prophet model
    model = Prophet(seasonality_mode='additive') # Start with additive seasonality
    model.fit(prophet_df)
    st.success("Prophet model fitted successfully!")

    # User input for forecast periods
    periods = st.number_input('Number of future periods to forecast:', min_value=1, value=30, step=1)
    freq = st.selectbox('Frequency of forecast periods:', options=['D', 'W', 'M', 'Y'], index=0, help='D: Daily, W: Weekly, M: Monthly, Y: Yearly')

    # Create future DataFrame for predictions
    future = model.make_future_dataframe(periods=int(periods), freq=freq)

    # Generate predictions
    forecast = model.predict(future)

    st.subheader("Forecast Data Preview")
    st.dataframe(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())

    # Store forecast data and model in session_state for visualization in the next step
    st.session_state['prophet_model'] = model
    st.session_state['forecast_data'] = forecast
    st.session_state['df_for_prophet_display'] = prophet_df # Store the prepared df for plotting

    st.success("Forecasts generated and stored for visualization!")

else:
    st.warning("Please ensure a dataset is uploaded and time/target columns are selected in previous steps.")

# To ensure global_df is consistent with the latest processed data for subsequent cells in Colab
if 'processed_df_for_ts' in st.session_state and not st.session_state['processed_df_for_ts'].empty:
    global_df = st.session_state['processed_df_for_ts']
else:
    global_df = pd.DataFrame()
```
